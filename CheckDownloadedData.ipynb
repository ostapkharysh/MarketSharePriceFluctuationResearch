```python
step= datetime.timedelta(minutes=15)
```

```python
count = 0
av_data = []
sh_data=[]
for el in os.listdir("//media/ostapkharysh/SP_PHD_U3/gdelt/"):
    #if el.startswith('2016') or el.startswith('2017'):
    el = el.split(".")[0]
    data = datetime.datetime.strptime(el, '%Y%m%d%H%M%S')
    av_data.append(data)
    count+=1
print(count)
```

```python
day = datetime.datetime(year=2016,month=1,day=1,hour=0,minute=0)
while day.year!=2019:
    sh_data.append(day)
    day+=step
```

```python
to_download_data = []
for el in sh_data:
    if el not in av_data:
        to_download_data.append(el.strftime('%Y%m%d%H%M%S'))
print(to_download_data)
```
