```python
def conv_to_date(el):
    el = el.split('T')
    print(el)
    el = el[0]+" "+el[1].replace('Z', '')
    return datetime.strptime(el, '%Y-%m-%d %H:%M:%S')
    
```

```python
df.head()
```

```python
df['time'] = [conv_to_date(x) for x in df['time']]
```

```python
df = df.sort_values(by='time')
```
