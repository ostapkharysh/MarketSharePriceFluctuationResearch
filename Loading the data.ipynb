```python
import pandas as pd
from datetime import datetime
```

```python
#pd.read_csv('Google.csv')
#pd.DataFrame.sort_values('1')
#df.sort_values(df.columns[2])
df = pd.DataFrame(data=pd.read_csv('Google.csv'))
```

```python
def conv_to_date(el):
    el = el.split('T')
    el = el[0]+" "+el[1].replace("Z", "")
    return datetime.strptime(el, '%Y-%m-%d %H:%M:%S')
    
```

```python
df.head()
```

```python
df['time'] = [conv_to_date(x) for x in df['time']]
```

```python
df = df.sort_values(by='time')
```
