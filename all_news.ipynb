```python
from newsapi import NewsApiClient
import csv
newsapi = NewsApiClient(api_key='0763606ae56d49e08902365e0dbcb239')
```

```python
with open('Google.csv', 'a') as csvfile:
    fieldnames = ['link', 'time']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for i in range(1,10):
        all_articles = newsapi.get_everything(q='Google',
                                          #sources='bbc-news,the-verge',
                                          #domains='bbc.co.uk,techcrunch.com',
                                          from_param='2018-09-01',
                                          to='2018-09-29',
                                          language='en',
                                          sort_by='relevancy',
                                          page_size = 100,
                                          page=i)
        data = [[el['url'], el['publishedAt']] for el in all_articles['articles']]
        for el in data:
            writer.writerow({'link': el[0],
                         'time': el[1]})
    
    #with open('google.txt', 'a') as g:
    #    for el in data:
    #        g.write(el[0] +', '+el[1]+'\n')
```
